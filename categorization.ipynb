```python
import os
import numpy as np 
import pandas as pd
```

```python
# setting your own relative data_path for trainning set
dir_data = 'data_set'

f_train = os.path.join(dir_data, 'task1_trainset.csv')
# f_test = os.path.join(dir_data, 'task1_public_testset.csv')

# read file and convert into pandas dataframe
train_data = pd.read_csv(f_train)
# test_data = pd.read_csv(f_test)
```

```python
train_data.head()
```

```python
train_data['Abstract_Split'] = train_data['Abstract'].apply(lambda a : a.split('$$$'))
train_data[['Abstract', 'Abstract_Split']].head()
```

```python
train_data['Task_Split'] = train_data['Task 1'].apply(lambda t : t.split(' '))
train_data[['Task 1', 'Task_Split']].head()
```

```python
train_data['Categories_Split'] = train_data['Categories'].apply(lambda c : c.split('/'))
train_data[['Categories', 'Categories_Split']].head()
```

```python
train_data['Domain'] = train_data['Categories_Split'].apply \
    (lambda cs : [c.split('.')[0] for c in cs] )

train_data[['Categories_Split', 'Domain']].head()
```

```python
# Some categories name in the data set may not contain sub category
# eg. quant-ph in D01568
train_data['Sub_Categories'] = train_data['Categories_Split'].apply \
    (lambda cs : [c.split('.')[1] if len(c.split('.')) is 2 else None for c in cs] )
   
train_data[['Categories_Split', 'Sub_Categories']].head()
```

```python
train_data.head()
```

---

```python
sentence_task_dict = []

for index, row in train_data.iterrows():
    for sen, cate in zip(row['Abstract_Split'], row['Task_Split']):
        sentence_task_dict.append({
            'Sentence': sen,
            'Task': cate
        })

sentence_task_dict[:3]
```

```python
# using python dict to construct new dataframe
abstract_task = pd.DataFrame(sentence_task_dict)
abstract_task.head()
```

---

```python
sentence_domain_dict = []

for index, row in train_data.iterrows():
    for sen in row['Abstract_Split']:
        sentence_domain_dict.append({
            'Sentence': sen,
            'Domain': row['Domain']
        })
        
sentence_domain_dict[:3]
```

```python
# using python dict to construct new dataframe
abstract_domain = pd.DataFrame(sentence_domain_dict)
abstract_domain.head()
```

---

```python
sentence_sub_category_dict = []

for index, row in train_data.iterrows():
    for sen in row['Abstract_Split']:
        sentence_sub_category_dict.append({
            'Sentence': sen,
            'Sub_Categories': row['Sub_Categories']
        })
        
sentence_sub_category_dict[:3]
```

```python
# using python dict to construct new dataframe
abstract_sub_category = pd.DataFrame(sentence_sub_category_dict)
abstract_sub_category.head()
```
